```python
df = pd.read_csv('C:/Users/josep/OneDrive/Desktop/Excel Studies/datasets to clean/global_freelancers_raw.csv')
df
```

```python
df.isna().sum()
```

```python
print(df['gender'].unique())

df['gender'] = df['gender'].apply(lambda x: 'Female' if x.lower() in ('f','female') else 'Male')
df
```

```python
print(df['age'].unique())

df['age'] = df['age'].fillna('Not Specified')
df.loc[df['age'] != 'Not Specified', 'age'] = df.loc[df['age'] != 'Not Specified', 'age'].astype(int)
df
```

```python
print(df['years_of_experience'].unique())

df['years_of_experience'] = df['years_of_experience'].fillna('Not Specified')
df.loc[df['years_of_experience'] != 'Not Specified', 'years_of_experience'] = (
    df.loc[df['years_of_experience'] != 'Not Specified', 'years_of_experience'].astype(int)
)
df
```

```python
print(df['hourly_rate (USD)'].unique())

df['hourly_rate (USD)'] = df['hourly_rate (USD)'].fillna('Not Specified')
df['hourly_rate (USD)'] = (
    df['hourly_rate (USD)']
    .str.replace('USD ','')
    .str.replace('$','')
)
df
```

```python
print(df['rating'].unique())

df['rating'] = df['rating'].fillna('Not Specified')
df['rating'] = df['rating'].apply(lambda x: 'Not Specified' if x==0 else x)
df
```

```python
print(df['is_active'].unique())

df['is_active'] = df['is_active'].fillna('Not Specified')
df.loc[df['is_active'] != 'Not Specified', 'is_active'] = (
    df.loc[df['is_active'] != 'Not Specified', 'is_active'].apply(lambda x: True if x in ('1','yes','Y') else False)
)
df
```

```python
print(df['client_satisfaction'].unique())

df['client_satisfaction'] = df['client_satisfaction'].fillna('Not Specified')
df['client_satisfaction'] = df['client_satisfaction'].str.replace('%','')
df
```
